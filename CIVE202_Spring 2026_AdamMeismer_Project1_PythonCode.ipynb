Client Request #3

```python
def assign_temperature(X):
    if X < 32:
        return "Below Freezing"
    elif 32 <= X <= 50:
        return 'Cool'
    elif 51 <= X <= 70:
        return 'Warm'
    else:
        return 'Hot'
```

```python
def assign_humidity(Y):
    if Y < 50:
        return "Low Humidity"
    elif 50 <= Y <= 80:
        return 'High Humidity'
    else:
        return 'Very High Humidity'
```

```python
air_data_vocTEMP = air_data.groupby(['sensor.name','date','temperature']).agg(
    max=('voc','max')
)

top5_voc_maxTEMP = air_data_vocTEMP.sort_values(by='max', ascending=False)

top5_voc_maxTEMP = top5_voc_maxTEMP.head(5).reset_index()

top5_voc_maxTEMP['Temperature Level'] = \
    top5_voc_maxTEMP['temperature'].apply(assign_temperature)

top5_voc_maxTEMP
```

```python
air_data_vocHUM = air_data.groupby(['sensor.name', 'date', 'temperature']).agg(
    max=('voc', 'max')
)

top5_voc_maxHUM = air_data_vocHUM.sort_values(by='max', ascending=False)

top5_voc_maxHUM = top5_voc_maxHUM.reset_index()

top5_voc_maxHUM['Temperature Level'] = \
    top5_voc_maxHUM['temperature'].apply(assign_humidity)

top5_voc_maxHUM.head(5)
```

Client Prompt #4

```python
def assign_aqi_info(Z):
    if Z <= 50:
        return "Good", "Green"
    elif Z <= 100:
        return "Moderate", "Yellow"
    elif Z <= 150:
        return "Unhealthy for Sensitive Groups", "Orange"
    elif Z <= 200:
        return "Unhealthy", "Red"
    elif Z <= 300:
        return "Very Unhealthy", "Purple"
    else:
        return "Hazardous", "Maroon"
```

```python
def calc_aqi_from_breakpoints(C, breakpoints):
    """
    C = concentration
    breakpoints = list of tuples (Clow, Chigh, Ilow, Ihigh)
    """
    if pd.isna(C):
        return None

    for Clow, Chigh, Ilow, Ihigh in breakpoints:
        if Clow <= C <= Chigh:
            return (Ihigh - Ilow) / (Chigh - Clow) * (C - Clow) + Ilow

    if C > breakpoints[-1][1]:
        return 500

    return None

pm25_bps = [
    (0.0,   12.0,   0,   50),
    (12.1,  35.4,  51,  100),
    (35.5,  55.4, 101,  150),
    (55.5, 150.4, 151,  200),
    (150.5,250.4, 201,  300),
    (250.5,350.4, 301,  400),
    (350.5,500.4, 401,  500),
]

pm10_bps = [
    (0,   54,   0,   50),
    (55, 154,  51,  100),
    (155,254, 101,  150),
    (255,354, 151,  200),
    (355,424, 201,  300),
    (425,504, 301,  400),
    (505,604, 401,  500),
]
air_data["pm2.5_atm"]  = pd.to_numeric(air_data["pm2.5_atm"], errors="coerce")
air_data["pm10.0_atm"] = pd.to_numeric(air_data["pm10.0_atm"], errors="coerce")
air_data["date"]       = pd.to_datetime(air_data["date"], errors="coerce")

air_data["AQI_PM25"] = air_data["pm2.5_atm"].apply(lambda C: calc_aqi_from_breakpoints(C, pm25_bps))
air_data["AQI_PM10"] = air_data["pm10.0_atm"].apply(lambda C: calc_aqi_from_breakpoints(C, pm10_bps))

air_data_AQI25 = air_data.groupby(["sensor.name", "date"]).agg(
    max=("AQI_PM25", "max")
)

top5_AQI25 = air_data_AQI25.sort_values(by="max", ascending=False)
top5_AQI25 = top5_AQI25[top5_AQI25["max"] >= 101].reset_index()

top5_AQI25["AQI Category"] = top5_AQI25["max"].apply(assign_aqi_info)
top5_AQI25["AQI Color"]    = top5_AQI25["max"].apply(assign_aqi_info)

top5_AQI25
air_data_AQI10 = air_data.groupby(["sensor.name", "date"]).agg(
    max=("AQI_PM10", "max")
)

top5_AQI10 = air_data_AQI10.sort_values(by="max", ascending=False)
top5_AQI10 = top5_AQI10[top5_AQI10["max"] >= 101].reset_index()

top5_AQI10["AQI Category"] = top5_AQI10["max"].apply(assign_aqi_info)
top5_AQI10["AQI Color"]    = top5_AQI10["max"].apply(assign_aqi_info)

top5_AQI10
```
